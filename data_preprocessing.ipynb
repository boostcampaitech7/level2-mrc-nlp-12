```python
import re
import pandas as pd

# 데이터 로드 (JSON 파일로부터 데이터 로드)
data = pd.read_json("./data/wikipedia_documents_unique.json")

def preprocess(text):
    text = re.sub(r"\n", " ", text)
    text = re.sub(r"\\n", " ", text)
    text = re.sub(r"#", " ", text)
    text = re.sub(r"[^A-Za-z0-9가-힣.?!,()~‘’“”"":%&《》〈〉''㈜·\-\'+\s一-龥サマーン]", "", text)  # サマーン 는 predictions.json에 있었음
    text = re.sub(r"\s+", " ", text).strip()  # 두 개 이상의 연속된 공백을 하나로 치환
    
    return text

# 전처리 적용
data["text"] = data["text"].apply(preprocess)

# 원래 구조 유지하며 JSON 파일로 저장 (다시 transpose 하지 않음)
data.transpose().to_json("./data/preprocessed_wiki_docs_unique.json", orient = "columns", force_ascii = False, indent = 4)
```

AEDA(Reader)

```python
from datasets import load_from_disk
import pandas as pd

dataset = load_from_disk("./data/train_dataset/")
train_dataset = dataset["train"]
```

```python
import random
from datasets import Dataset, concatenate_datasets

def apply_aeda(text, punc_list = None):
    """
    AEDA를 적용하는 함수. 임의의 구두점을 문장에 삽입하여 데이터를 증강함.
    text: 원본 문장
    punc_list: 삽입할 구두점 리스트 (기본값으로 , . ? ! : ; 적용)
    """
    if punc_list is None:
        punc_list = [",", ".", "?", "!", ":", ";"]
    
    words = text.split()
    num_punc_insertions = random.randint(1, max(1, len(words) // 3))
    
    for _ in range(num_punc_insertions):
        insert_position = random.randint(0, len(words) - 1)
        punc = random.choice(punc_list)
        words.insert(insert_position, punc)
    
    return " ".join(words)

def augment_dataset_with_aeda(dataset, augment_ratio = 0.3):
    """
    ODQA 데이터셋에 대해 AEDA를 적용하여 증강함.
    dataset: KorQuAD와 유사한 ODQA 데이터셋
    augment_ratio: 증강할 데이터 비율
    """
    augmented_data = []
    num_examples = len(dataset)
    num_augment = int(num_examples * augment_ratio)

    selected_indices = random.sample(range(num_examples), num_augment)
    
    for idx in selected_indices:
        original_question = dataset[idx]["question"]
        augmented_question = apply_aeda(original_question)
        
        new_example = {
            "title": dataset[idx]["title"],
            "context": dataset[idx]["context"],
            "question": augmented_question,
            "id": dataset[idx]["id"] + "_aug",  # id에 "_aug" 추가하여 구분
            "answers": dataset[idx]["answers"],  # 답변은 동일하게 유지
            "document_id": dataset[idx]["document_id"]
        }
        augmented_data.append(new_example)
    
    # 증강된 데이터를 Huggingface Dataset으로 변환
    augmented_dataset = Dataset.from_pandas(pd.DataFrame(augmented_data))
    
    # 기존 Arrow 데이터셋과 병합 (concatenate)
    combined_dataset = concatenate_datasets([dataset, augmented_dataset])
    
    return combined_dataset

# 질문을 30% 비율로 증강
augmented_train_dataset = augment_dataset_with_aeda(train_dataset, augment_ratio = 0.3)

augmented_train_dataset.save_to_disk("./data/aeda_train_dataset")
```

AEDA(Retriever)

```python
import random
import pandas as pd
import json

def apply_aeda(text, punc_list = None):
    """
    AEDA를 적용하는 함수. 임의의 구두점을 문장에 삽입하여 데이터를 증강함.
    text: 원본 문장
    punc_list: 삽입할 구두점 리스트 (기본값으로 , . ? ! : ; 적용)
    """
    if punc_list is None:
        punc_list = [",", ".", "?", "!", ":", ";"]
    
    words = text.split()
    num_punc_insertions = random.randint(1, max(1, len(words) // 3))
    
    for _ in range(num_punc_insertions):
        insert_position = random.randint(0, len(words) - 1)
        punc = random.choice(punc_list)
        words.insert(insert_position, punc)
    
    return " ".join(words)

def augment_wikipedia_data(dataframe, augment_ratio = 0.3):
    """
    Wikipedia 데이터에서 "text" 필드에 대해 AEDA를 적용하여 증강함.
    dataframe: Wikipedia 데이터가 포함된 pandas DataFrame
    augment_ratio: 데이터 증강 비율
    """
    augmented_data = []
    num_examples = len(dataframe)
    num_augment = int(num_examples * augment_ratio)
    
    # 무작위로 선택된 데이터 인덱스 추출
    selected_indices = random.sample(range(num_examples), num_augment)
    
    for idx in selected_indices:
        original_text = dataframe.loc[idx, "text"]
        augmented_text = apply_aeda(original_text)
        
        new_row = {
            "text": augmented_text,
            "corpus_source": dataframe.loc[idx, "corpus_source"],
            "url": dataframe.loc[idx, "url"],
            "domain": dataframe.loc[idx, "domain"],
            "title": dataframe.loc[idx, "title"],
            "author": dataframe.loc[idx, "author"],  # 추가된 열들 포함
            "html": dataframe.loc[idx, "html"],
            "document_id": str(dataframe.loc[idx, "document_id"]) + "_aug"
        }
        augmented_data.append(new_row)
    
    # 증강된 데이터를 DataFrame으로 변환
    augmented_dataframe = pd.DataFrame(augmented_data)
    
    # 원본 데이터와 증강된 데이터를 병합
    combined_dataframe = pd.concat([dataframe, augmented_dataframe], ignore_index = True)
    
    return combined_dataframe

# JSON 파일에서 데이터 읽기
df = pd.read_json("./data/wikipedia_documents.json")

# 데이터 전치 (추가적인 열들이 나타나는 경우 전치 사용)
df = df.T

# AEDA를 30% 비율로 적용하여 증강
augmented_df = augment_wikipedia_data(df, augment_ratio = 0.3)

# 데이터 다시 전치해서 원래 형태로 복원
augmented_df = augmented_df.T

# 증강된 데이터를 JSON 파일로 저장
augmented_df.to_json("./data/augmented_wikipedia_documents.json", orient = "columns", force_ascii = False, indent = 4)
```
